# Part 1: Rank
The rank of a matrix is the maximum number of linearly independent rows or columns it contains, representing the dimension of its row or column space. It indicates the matrix's information content, with full rank meaning maximum possible independence.


*   The dimension of the row (or column) space of a matrix $A$ is the rank of $A$ and is denoted by $rank(A)$
*   The row space and column space of an $m$ x $n$ matrix $A$ have the same dimensions.
*   For an $m$ x $n$ matrix, the rank is at most $min(m, n)$.
*   An $m$ x $n$ matrix has full rank if its rank equals $min(m, n)$.
*   The Rank-Nullity Theorem states that for an $n$-column matrix: $rank(A) + Null(A) = n$





```python
# Features: house size, rooms, lot size(redundant)

X = np.array([
    [1000, 3, 2000],
    [1500, 4, 3000],
    [2000, 5, 4000],
    [2500, 6, 5000]
])

print("Shape (m, n):", X.shape)
print("Rank:", np.linalg.matrix_rank(X))
```

# Fit a Linear Model

The NumPy function `np.linalg.lstsq(feature_matrix, target)` solves the least squares problem by computing the best-fit solution to a linear system and minimizing the squared error. It is used in linear regression when there is no exact solution or the system is overdetermined.

In a real dataset, we usually have more observations than the number of features. Therefore, no exact solution fits every data point perfectly, so `np.linalg.lstsq(feature_matrix, target)` funds the coefficients that minimize the overall squaredd error.

**Understand the Function**
`np.linalg.lstsq(feature_matrix, target)`

*   `feature_matrix`: matrix of the study (shape: observations x features)
*   `target`: target vector (shape: observation)

**Output:**


*   **Solution**: weights or scalar in the linear combination of the system
*   **Residuals**: sum of squared errors
*   **Rank**: rank of `feature_matrix`
*   **Singular value**: factor that the matrix stretches or scales data in different directions, is used for stability

    *   Large singular value indicates strong, important direction in the data
    *   Small singular value indicates weak, nearly redundant direction
    *   Zero (or close to zero) means no information in that direction
    *   For small singular values, `np.linalg.lstsq(feature_matrix, target)` reduces the weight of a direction when computing the solution.









```python
# Target vector (yearly lease)
y = np.array([18000, 28000, 38000, 48000])

# Compute the least-square solution to the linear matrix equation
w, residuals, rank, s = np.linalg.lstsq(X_new, y)

print("Weights:", w)
print("Residuals:", residuals)
print("Rank:", rank)
print("Singular values:", s)
```
