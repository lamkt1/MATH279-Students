# Exercise 3: Verifying Vector Space Properties (3 points)

Let $V$ be a set on which two operations (**vector addition** and **scalar multiplication**) are defined. If the listed axioms are satisfied for every **$u$**, **$v$**, **$w$** in $V$ and every scalar *c* and *d*, then $V$ is a **vector space**.

**Vector Addition**

1.   **Closure under addition:** ${u}+{v}$ is in $V$
2.   **Commuatative property:** ${u}+{v}={v}+{u}$
3.   **Associativate property:**$({u}+{v})+{w}={u}+({v}+{w})$
4.   **Additive identity:** $V$ has a zero vector $0$ such that for every ${u}$ in $V$, ${u}+{0}={u}$
5.   **Additive inverse:**For each ${u}$ in $V$, there is a vector $-{u}$ in $V$ such that ${u}+(-{u})={0}$

**Scalar Multiplication**

1.   **Closure under scalar multiplication:** $c{u}$ is in $V$
2.   **Distributive property:** $c({u}+{v})=c{u}+c{v}$
3.   **Distributive property:** $(c+d){u}=c {u}+d{u}$
4. **Associativate property:**$ c(d{u})=(c d){u}$
5. **Scalar identity:** $1{u}={u}$
