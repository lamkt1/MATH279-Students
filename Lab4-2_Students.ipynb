# Part 1-2: Linear Independence

Linear independence means no vector in the set can be written as a linear combination of the others. In other words, no redundancy. They are independent if the only solution to $$v =c_{1} {v}_{1}+c_{2} {v}_{2}+\cdots+c_{n} {v}_{n} = 0$$ is $c_{1} = c_{2} = \cdots =c_{n}= 0$, which is called the trivial solution.

Geometrically, these vectors represent unique directions that cannot be formed by combining others, crucial for forming a basis.

A set of $n$ linearly independent vectors in an $n$-dimensional space forms a basis, meaning they can span the entire space without redundant information.

For $c_{1} {v}_{1}+c_{2} {v}_{2}=0$,


*   If nonzero solution exists → dependent
*   Only zero solution → independent

# Part 3: Prinical Component Analysis

Principal Component Analysis (PCA) is a foundational unsupervised learning and dimensionality reduction technique used in statistics, data science, and machine learning. It transforms a set of possibly correlated variables into a smaller number of uncorrelated variables called principal components, while preserving as much data variability as possible.

Datasets often have redundant features and high dimensionality, which makes analysis, visualization, and modeling difficult.

PCA can:


*   transform dimensionality efficiently
*   remove multicollinearity
*   improve computational efficiency
*   no labels required (unsupervised learning)
*   discover hidden patterns



```python
# Approach 1: Learn about .fit() and .transform()
# Original data (with redundancy)
x1 = np.array([1, 2, 3, 4, 5])
x2 = 2 * x1
x3 = np.array([2, 3, -1, 5, 8])

# Stack x1, x2, x3 as columns
X = np.column_stack((x1, x2, x3))
print("Original data:\n", X)

# Learning data
model = PCA(n_components=2)
model.fit(X)

# Apply PCA and transform the learning patterns to optimal rotation of data
X_transform = model.transform(X)

print("Transformed data:\n", X_transform)
```

```python
# Approach 2: Learn about .fit_transform()
# Original data (with redundancy)
x1 = np.array([1, 2, 3, 4, 5])
x2 = 2 * x1
x3 = np.array([2, 3, -1, 5, 8])

# Stack x1, x2, x3 as columns
X = np.column_stack((x1, x2, x3))
print("Original data:\n", X)

model = PCA(n_components=2)

X_transform2 = model.fit_transform(X)

print("Other transformed data result:\n", X_transform2)
```
